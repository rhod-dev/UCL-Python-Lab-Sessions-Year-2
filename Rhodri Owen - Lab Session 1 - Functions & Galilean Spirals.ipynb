## Introduction
This is a Jupyter notebook to calculate and plot some functions. First, we input the numpy and plotting libraries as follows:

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Conversion of polar to cartesian co-ordinates
First, we write a function to convert polar co-ordaintes (given in terms of $r$ and $\theta$) to Cartesian co-ordinates $x$ and $y$. We use the definition of polar co-ordinates to give the conversion:
cartesian coordinates $x,y$, using the equations 
$$ x = r \cos(\theta) $$
$$ y = r \sin(\theta) $$
We use numpy as the library for functions such as sine and cosine.

```python
def polar_to_cartesian_fn(r,theta):
    '''Converts polar co-ordinates to cartesian co-ordinates. Input r, theta; output x, y'''
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    return x,y
```

We then set arbitrary values of r and theta and check that the function works as intended:

```python
r = 5
theta = np.pi/8
x,y = polar_to_cartesian_fn(1,np.pi/2)
print(x,y)
```

### Plotting the Galilean spiral
We then want to plot the Galilean spiral given by $r = \theta^2 - l$, in the range $0<\theta<10\pi$, interval $0.1$.  First, we create an array with the given inputs of theta. Then, we calcualte the set of values of r, and convert these to Cartesian co-ordinates. Finally, we plot these data using matplotlib.

```python
interval = 0.1
theta_in = np.arange(0,10*np.pi,0.1) # creates an array of values for theta
r_out = theta_in**2 - interval # calculates corresponding values of r

x,y = polar_to_cartesian_fn(r_out,theta_in) # converts to cartesian co-ordinates
```

```python
# plotting the galilean spiral with matplotlib
plt.plot(x,y, 'b-', label="Galilean spiral for l= 0.1")
plt.title('The Galilean Spiral')
plt.xlabel("x")
plt.ylabel("y")
```

### Variation of the Interval
We then create an array of values for l, to see how changing the values changes the spiral. We use a for loop to calculate the co-ordinates for each of these values of l. Finally, we plot these and give the graph a legend.

```python
l_array = np.linspace(0,10,5)
theta_in = np.linspace(0,2*np.pi,100)
for i in range(len(l_array)):
    r_out=theta_in**2 - l_array[i] # Calculate r
    x,y = polar_to_cartesian_fn(r_out,theta_in) # Convert to cartesian co-ords
    plt.plot(x,y, label="$l= {0:3.1f}$".format(l_array[i]))# Plot on axes

plt.title('The Galilean Spiral for different values of l')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(loc=0)# 0 chooses the best location for the legend
```
